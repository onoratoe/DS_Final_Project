```python
import pandas as pd
import seaborn as sns
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('parsed_url_features.csv')
df.head(10)
```

```python
# Transform the data
scaler = StandardScaler()
# transform numerical values only
df_scaled = scaler.fit_transform(df.iloc[:,12:])
df_scaled = pd.DataFrame(df_scaled, columns=df.columns[12:])
# drop null values
df_scaled = df_scaled.dropna()
df_scaled.head(10)
```

```python
# Visualize the numerical data with Exploratory Data Analysis
# Pairplot of the numerical data, with 'status' 
for col in df.columns[12:]:
    plt.figure(figsize=(10, 6))
    sns.pairplot(df_scaled, x_vars=[col], y_vars=['status'], height=5, aspect=1)
    plt.title(f'{col} vs status')
    plt.xlabel(col)
    plt.ylabel('status')
    plt.show()
    
```
